```python
# This is a recreation of how the professor cleaned his data, which we will use a guide for further cleaning
import pandas as pd
```

```python
# Loading the CSV data. row 0 and row 1 as header.
# This creates a multi index
df = pd.read_csv('~/Desktop/kenvue_data/Customer DC Inventory - UTSC Lecture.csv', header=[0,1])
```

```python
# Let's see what it looks like
df
```

```python
# This is called multi index. 
df.columns
```

```python
# Instead of having numbers in row index, we need to set customized index levels
df = df.set_index([('Fiscal Year  /  Fiscal Week', 'Need State'), ('Unnamed: 1_level_0', 'Unnamed: 1_level_1')])

df.columns
```

```python
df
```

```python
# Swap rows and columns
df = df.T
```

```python
# Looking good 
df
```

```python
# we use a Pandas stack function here to stack all the need states in one column.  
df = df.stack(0)
```

```python
# Can you see what I did so far?
df
```

```python
# Now I will reset the index so it will go back to numbers as row index
df= df.reset_index()
```

```python
df
```

```python
# Let's rename the columns
df = df.rename_axis(None, axis=1)
df = df.rename(columns={df.columns[0]: "year", df.columns[1]: "week", df.columns[2]: "Need States"})
```

```python
df
```

```python
# Here I use the datetime function to convert the week# to an acutual date for easy plotting
df['date'] = pd.to_datetime(df["year"].astype(str) + "-" + df["week"].astype(str) + "-1", format='%Y-%U-%w')
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.lineplot(data = df, x=df.date, y='Dc Amount')
```

```python
sns.lineplot(data = df, x='week', y='Dc Amount', hue='year')
```

```python
# This is the code to reorganize the Factory POS$ csv provided by kenvue
import pandas as pd
```

```python
# Loading the CSV data. I declared row 0 as header.  The data begins in row 1.
df = pd.read_csv('~/Desktop/kenvue_data/Factory POS $ - UTSC Lecture.csv', header=[0,1])
```

```python
# Let's see what it looks like
df
```

```python

df.columns
```

```python
#Need to set the index to be the Fiscal year/Fiscal week column since only the needstates need to be stacked
df2 = df.set_index([('Unnamed: 0_level_0', 'Need State'), ('Unnamed: 1_level_0', 'Fiscal Week')])
```

```python
df2
```

```python
df2 = df2.T
```

```python
df2
```

```python

df2 = df2.stack(1)
```

```python
df2
```

```python

df2 = df2.reset_index()
```

```python

df2
```

```python
df2 = df2.drop('level_0', axis=1)
```

```python
df2
```

```python
# Let's rename the columns to match the rest of the data we cleaned
df2 = df2.rename_axis(None, axis=1)
df2 = df2.rename(columns={df2.columns[0]: "year", df2.columns[1]: "week"})
```

```python

df2
```

```python
df2 = df2.set_index([('year'),('week')])
```

```python
df2
```

```python
df2 = df2.stack(0)
```

```python
df2 = df2.reset_index()
```

```python
df2
```

```python
# Let's rename the columns to match the rest of the data we cleaned
df2 = df2.rename_axis(None, axis=1)
df2 = df2.rename(columns={df2.columns[2]: "Need States", df2.columns[3]: "Factory POS"})
```

```python
# Lastly, use the datetime function to add a date column that resembles excel better
df2['date'] = pd.to_datetime(df2["year"].astype(str) + "-" + df2["week"].astype(str) + "-1", format='%Y-%U-%w')
```
